```python
import pandas as pd
import numpy as np
```

```python
mempool=pd.read_csv("mempool.csv")
```

## Printing the dataset

```python
mempool
```

## Getting total fee and total weight

```python
mempool=mempool.append(mempool[['fee','weight']].sum(),ignore_index=True)
mempool
```

## Printing out all the rows for which parent is Nan

```python
mempool_nanval=mempool[mempool['parents '].isna()]
mempool_nanval
```

## Getting the number of **Nan** values in parents column

```python
Nan_count=mempool['parents '].isna().sum()
Nan_count
```

## Checking for the values having the same Parent

```python
dups = mempool.pivot_table(index = ['parents '], aggfunc ='size')
dups
```

## Droping the duplicate values so that we can get the total number of values having single parent

```python
mempool.drop_duplicates(subset ="parents ",
                     keep = False, inplace = True)
mempool
```

```python
Single_parent_count=len(mempool['parents '].value_counts())
Single_parent_count
```

```python
TotalTransaction_count=Single_parent_count+Nan_count
TotalTransaction_count
```

## Dataframe after cleaning. This Dataframe contains the Final Transactions Information

```python
mempool_concatated=pd.concat([mempool_nanval,mempool])
mempool_concatated
```

## This is the Final Transactions Dataframe

```python
Final_transaction=mempool_concatated.drop(['fee','weight','parents '],axis=1)
Final_transaction
```

```python
Final_transaction.to_csv("Block.txt", header=False, index=False, sep='\t', mode='a')
```
